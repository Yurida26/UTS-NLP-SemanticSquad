Notebook ini mencakup:
1. **Data Loading:** Mengambil data mentah ulasan produk Tokopedia.
2. **Random Sampling:** Mengambil sampel data secara acak untuk efisiensi komputasi.
3. **Data Balancing:** Menyeimbangkan jumlah data per kategori agar model tidak bias.
4. **Export Data:** Menyimpan dataset yang sudah seimbang untuk tahap preprocessing.

**IMPORT & LOAD**

Memuat Dataset Utama

```python
import pandas as pd

# Upload file ke folder Colab
file_path = 'tokopedia_product_reviews_2025.csv'

try:
    df = pd.read_csv(file_path)
    print("Dataset utama berhasil dimuat!")
    print(f"Total data asli: {len(df)} baris")
except:
    print("File tidak ditemukan! Pastikan file CSV sudah di-upload ke folder Colab (ikon folder di kiri).")
```

**SAMPLING**

Pengambilan Sampel Acak & Penyeimbangan

```python
# Mengambil maksimal 2000 data per kategori supaya seimbang
# random_state=42 ditambahkan agar hasil sampling selalu SAMA setiap kali di-run
df_balanced = df.groupby('product_category').apply(
    lambda x: x.sample(n=min(len(x), 2000), random_state=42)
).reset_index(drop=True)

# Cek hasil penyeimbangan
print("Jumlah data per kategori (Balanced):")
print(df_balanced['product_category'].value_counts())
print(f"\nTotal data keseluruhan: {len(df_balanced)}")

# Simpan jadi file baru untuk tahap Preprocessing
df_balanced.to_csv('dataset_uts_12k_balanced.csv', index=False)
print("\nFile 'dataset_uts_12k_balanced.csv' sukses dibuat!")
```
